```python
import pandas as pd

# Load master produk Indofood
df_master = pd.read_excel("SKU.xlsx", sheet_name="All NIVEA PL Mei 2024")

# Load hasil OCR
df_ocr = pd.read_excel("SKUT.xlsx", sheet_name="Results")

# Join berdasarkan barcode
df_join = pd.merge(
    df_ocr,
    df_master,
    on="barcode_pieces",   # kolom kunci join
    how="left"             # ambil semua data OCR, lengkapi dari master
)

# Hitung total harga (qty * harga_jual_saran)
df_join["total_harga"] = df_join["qty"] * df_join["harga_jual_saran"]

# Simpan hasil join ke Excel baru
df_join.to_excel("OCR_Join_Result.xlsx", index=False)

print("Join selesai! Hasil disimpan ke OCR_Join_Result.xlsx")
```
